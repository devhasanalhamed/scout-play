```python
import os
import cv2

video_dataset_path = "./dataset"      # â† Ù…Ø¬Ù„Ø¯ Ø§Ù„ÙÙŠØ¯ÙŠÙˆÙ‡Ø§Øª Ø§Ù„Ù…ØµÙ†ÙØ©
output_image_path = "./frames_local"  # â† Ù…Ø¬Ù„Ø¯ Ø¬Ø¯ÙŠØ¯ Ù„Ø­ÙØ¸ Ø§Ù„ØµÙˆØ±
frame_interval = 30                   # ÙƒÙ„ 30 Ø¥Ø·Ø§Ø± = ØªÙ‚Ø±ÙŠØ¨Ù‹Ø§ ÙƒÙ„ Ø«Ø§Ù†ÙŠØ©

os.makedirs(output_image_path, exist_ok=True)

for category in os.listdir(video_dataset_path):
    category_path = os.path.join(video_dataset_path, category)
    if not os.path.isdir(category_path):
        continue

    category_output_path = os.path.join(output_image_path, category)
    os.makedirs(category_output_path, exist_ok=True)

    for video_name in os.listdir(category_path):
        if not video_name.lower().endswith((".mp4", ".mov", ".avi")):
            continue

        video_path = os.path.join(category_path, video_name)
        cap = cv2.VideoCapture(video_path)
        count = 0
        frame_num = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if frame_num % frame_interval == 0:
                img_name = f"{category}_{video_name[:-4]}_f{frame_num}.jpg"
                img_path = os.path.join(category_output_path, img_name)
                cv2.imwrite(img_path, frame)
                count += 1
            frame_num += 1

        cap.release()
        print(f" ØªÙ… Ø§Ø³ØªØ®Ø±Ø§Ø¬ {count} ØµÙˆØ±Ø© Ù…Ù† {video_name} â† ({category})")

print(" ØªÙ… Ø§Ù„Ø§Ù†ØªÙ‡Ø§Ø¡ Ù…Ù† Ø§Ø³ØªØ®Ø±Ø§Ø¬ ÙƒÙ„ Ø§Ù„Ø¥Ø·Ø§Ø±Ø§Øª.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Ø¥Ø¹Ø¯Ø§Ø¯ Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª
data_dir = "./frames_local"
img_height = 128
img_width = 128
batch_size = 32

# â¬… 1. ØªÙ‚Ø³ÙŠÙ… Ø§Ù„ØµÙˆØ± Ø¥Ù„Ù‰ ØªØ¯Ø±ÙŠØ¨ ÙˆØ§Ø®ØªØ¨Ø§Ø± (80/20)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% Ø§Ø®ØªØ¨Ø§Ø±
)

# â¬… 2. ØªØ­Ù…ÙŠÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨
train_data = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# â¬… 3. ØªØ­Ù…ÙŠÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ­Ù‚Ù‚ (Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±)
val_data = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)
```

```python
#  ScoutPlay - Ù…Ø±ÙƒØ² Ø§Ù„Ù„Ø§Ø¹Ø¨ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… Transfer Learning
# ------------------------------------------------------

#  1. ØªØ«Ø¨ÙŠØª TensorFlow
!pip install -q tensorflow

#  2. Ø§Ø³ØªÙŠØ±Ø§Ø¯ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
import os

#  3. ØªØ­Ù…ÙŠÙ„ Ø§Ù„ØµÙˆØ± Ù…Ù† Ù…Ø¬Ù„Ø¯ frames_local (Ù…Ø±ØªØ¨ Ø­Ø³Ø¨ Ø§Ù„ÙØ¦Ø§Øª)
data_dir = "./frames_local"  # â† ØªØ£ÙƒØ¯ÙŠ Ø¥Ù† Ø§Ù„ØµÙˆØ± Ù…ØµÙ†ÙØ© Ø¯Ø§Ø®Ù„ Ù…Ø¬Ù„Ø¯Ø§Øª: Ù…Ù‡Ø§Ø¬Ù… / ÙˆØ³Ø· / Ø¯ÙØ§Ø¹
img_height = 128
img_width = 128
batch_size = 32

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

#  4. Ø¨Ù†Ø§Ø¡ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… MobileNetV2
base_model = MobileNetV2(input_shape=(img_height, img_width, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # ØªØ¬Ù…ÙŠØ¯ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ©

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(3, activation='softmax')  # â† Ø¹Ø¯Ø¯ Ø§Ù„ÙØ¦Ø§Øª = 3
])

#  5. ØªØ¬Ù…ÙŠØ¹ ÙˆØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.compile(optimizer=Adam(learning_rate=0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_data,
    epochs=15,
    validation_data=val_data
)

#  6. Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.save("player_position_classifier.h5")
print(" ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ù†Ø¬Ø§Ø­!")

#  7. Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„ØªÙ†Ø¨Ø¤ Ø¹Ù„Ù‰ ØµÙˆØ±Ø© Ø¬Ø¯ÙŠØ¯Ø© (Ø§Ø®ØªÙŠØ§Ø±ÙŠ)
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    pred = model.predict(img_array)[0]
    classes = list(train_data.class_indices.keys())
    predicted_class = classes[np.argmax(pred)]
    confidence = np.max(pred)
    print(f" Ø§Ù„ØªÙˆÙ‚Ø¹: {predicted_class} ({confidence*100:.2f}%)")

#  Ø§Ø³ØªØ®Ø¯Ø§Ù…:
# predict_image("/content/frames_local/ÙˆØ³Ø·/ÙˆØ³Ø·_player1_f30.jpg")
```
