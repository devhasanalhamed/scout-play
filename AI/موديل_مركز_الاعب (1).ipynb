```python
from google.colab import files
import os
import shutil

# 👣 1. إنشاء مجلد المراكز
base_path = "/content/dataset"
categories = ["مهاجم", "وسط", "دفاع"]
for category in categories:
    os.makedirs(os.path.join(base_path, category), exist_ok=True)

# 👣 2. رفع فيديوهات كل مركز
for category in categories:
    print(f" ارفعي الآن فيديوهات مركز: {category}")
    uploaded = files.upload()  # ← هنا ترفعين الفيديوهات لهذا المركز

    for filename in uploaded.keys():
        shutil.move(filename, os.path.join(base_path, category, filename))

    print(f" تم حفظ الفيديوهات في: /dataset/{category}\n")
```

```python
import os
import cv2

video_dataset_path = "./dataset"      # ← مجلد الفيديوهات المصنفة
output_image_path = "./frames_local"  # ← مجلد جديد لحفظ الصور
frame_interval = 30                   # كل 30 إطار = تقريبًا كل ثانية

os.makedirs(output_image_path, exist_ok=True)

for category in os.listdir(video_dataset_path):
    category_path = os.path.join(video_dataset_path, category)
    if not os.path.isdir(category_path):
        continue

    category_output_path = os.path.join(output_image_path, category)
    os.makedirs(category_output_path, exist_ok=True)

    for video_name in os.listdir(category_path):
        if not video_name.lower().endswith((".mp4", ".mov", ".avi")):
            continue

        video_path = os.path.join(category_path, video_name)
        cap = cv2.VideoCapture(video_path)
        count = 0
        frame_num = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if frame_num % frame_interval == 0:
                img_name = f"{category}_{video_name[:-4]}_f{frame_num}.jpg"
                img_path = os.path.join(category_output_path, img_name)
                cv2.imwrite(img_path, frame)
                count += 1
            frame_num += 1

        cap.release()
        print(f" تم استخراج {count} صورة من {video_name} ← ({category})")

print(" تم الانتهاء من استخراج كل الإطارات.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# إعداد المسارات
data_dir = "./frames_local"
img_height = 128
img_width = 128
batch_size = 32

# ⬅ 1. تقسيم الصور إلى تدريب واختبار (80/20)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% اختبار
)

# ⬅ 2. تحميل بيانات التدريب
train_data = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

# ⬅ 3. تحميل بيانات التحقق (الاختبار)
val_data = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)
```

```python
#  ScoutPlay - مركز اللاعب باستخدام Transfer Learning
# ------------------------------------------------------

#  1. تثبيت TensorFlow
!pip install -q tensorflow

#  2. استيراد المكتبات
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
import os

#  3. تحميل الصور من مجلد frames_local (مرتب حسب الفئات)
data_dir = "./frames_local"  # ← تأكدي إن الصور مصنفة داخل مجلدات: مهاجم / وسط / دفاع
img_height = 128
img_width = 128
batch_size = 32

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

#  4. بناء النموذج باستخدام MobileNetV2
base_model = MobileNetV2(input_shape=(img_height, img_width, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # تجميد الطبقات الأساسية

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(3, activation='softmax')  # ← عدد الفئات = 3
])

#  5. تجميع وتدريب النموذج
model.compile(optimizer=Adam(learning_rate=0.0005),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_data,
    epochs=15,
    validation_data=val_data
)

#  6. حفظ النموذج
model.save("player_position_classifier.h5")
print(" تم حفظ النموذج بنجاح!")

#  7. اختبار التنبؤ على صورة جديدة (اختياري)
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    pred = model.predict(img_array)[0]
    classes = list(train_data.class_indices.keys())
    predicted_class = classes[np.argmax(pred)]
    confidence = np.max(pred)
    print(f" التوقع: {predicted_class} ({confidence*100:.2f}%)")

#  استخدام:
# predict_image("/content/frames_local/وسط/وسط_player1_f30.jpg")
```
